```python
import numpy as np
import pandas as pd
import cv2
```

```python
def get_png_from_mp4_secondwise(movieName, InDirectory="", OutDirectory=""):
    if OutDirectory != "":
        if OutDirectory[len(OutDirectory)-1] != "/":
            OutDirectory = OutDirectory + "/"
    video_capture = cv2.VideoCapture(InDirectory+movieName)
    success, image = video_capture.read()
    count    = 0
    old_time = -1
    time     = 0
    x1 = 168
    x2 = 180
    y1 = 300
    y2 = 325
    while success:
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        roi  = gray[y1:y2, x1:x2]
        time = video_capture.get(cv2.CAP_PROP_POS_MSEC)
        time = np.floor(time/1000)
        if time != old_time: # Event in next time (s) interval
            cv2.imwrite(OutDirectory+"image" + str(count) + ".png", roi)
        old_time = time
        success,image = video_capture.read()
        print('Read a new frame: ', success)
        count += 1
    return
```

```python
movieName = "TimoBollMaLong2017WC.mp4"
get_png_from_mp4_secondwise(movieName)
```

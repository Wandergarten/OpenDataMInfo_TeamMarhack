```python
import numpy as np
import pandas as pd
import cv2
```

```python
def get_png_from_mp4_secondwise_full_frame_resolution(movieName, InDirectory="", OutDirectory=""):
    if OutDirectory != "":
        if OutDirectory[len(OutDirectory)-1] != "/":
            OutDirectory = OutDirectory + "/"
    video_capture = cv2.VideoCapture(InDirectory+movieName)
    success, image = video_capture.read()
    count    = 0
    old_time = -1
    time     = 0
    x1 = 0
    x2 = 3000
    y1 = 0
    y2 = 3000
    image_counter = 0
    while success:
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        roi  = gray[y1:y2, x1:x2]
        time = video_capture.get(cv2.CAP_PROP_POS_MSEC)
        time = np.floor(time/1000)
        if time != old_time: # Event in next time (s) interval
            img_count = get_print_time_format(image_counter)
            cv2.imwrite(OutDirectory + "image" + img_count + ".png", roi)
            image_counter = image_counter + 1
        old_time = time
        success,image = video_capture.read()
        print('Read a new frame: ', success)
        count += 1
    return

def get_print_time_format(image_counter, length_numbering = 4):
    result = str(image_counter)
    num_zeros = length_numbering-len(result)
    for i in range(num_zeros):
        result = "0" + result
    return result
```

```python
# place the damn video in the 'Code' folder!
# if using a different video --> adapt pixel ratio
# it's best to create the folder structure beforehand manually
movieName = "WangHaoMaLin2008.mp4"
get_png_from_mp4_secondwise_full_frame_resolution(movieName, "", "./../Data/WangHaoMaLin2008_original/")
```

```python
def get_png_from_mp4_secondwise_halfed_added_white_space(movieName, InDirectory="", OutDirectory=""):
    if OutDirectory != "":
        if OutDirectory[len(OutDirectory)-1] != "/":
            OutDirectory = OutDirectory + "/"
    video_capture = cv2.VideoCapture(InDirectory+movieName)
    success, image = video_capture.read()
    count    = 0
    old_time = -1
    time     = 0
    x1 = 591
    x2 = 641
    y1 = 84
    y2 = 122
    a1 = 591
    a2 = 639
    b1 = 128
    b2 = 167
    image_counter = 0
    while success:
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        roi_upper  = gray[y1:y2, x1:x2]
        roi_lower  = gray[b1:b2, a1:a2]
        time = video_capture.get(cv2.CAP_PROP_POS_MSEC)
        time = np.floor(time/1000)
        bordersize = 10
        top = bordersize
        bottom = bordersize
        left = bordersize
        right = bordersize
        white = [255,255,255]
        roi_upper_added_white = cv2.copyMakeBorder(roi_upper, top, bottom, left, right, cv2.BORDER_CONSTANT, None, value = white)
        roi_lower_added_white = cv2.copyMakeBorder(roi_lower, top, bottom, left, right, cv2.BORDER_CONSTANT, None, value = white)
        if time != old_time: # Event in next time (s) interval
            img_count = get_print_time_format(image_counter)
            cv2.imwrite(OutDirectory + "image" + img_count + "_upper.png", roi_upper)
            cv2.imwrite(OutDirectory + "image" + img_count + "_lower.png", roi_lower)
            cv2.imwrite(OutDirectory + "image" + img_count + "_upper_white.png", roi_upper_added_white)
            cv2.imwrite(OutDirectory + "image" + img_count + "_lower_white.png", roi_lower_added_white)           
            image_counter = image_counter + 1
        old_time = time
        success,image = video_capture.read()
        print('Read a new frame: ', success)
        count += 1
    return

def get_print_time_format(image_counter, length_numbering = 4):
    result = str(image_counter)
    num_zeros = length_numbering-len(result)
    for i in range(num_zeros):
        result = "0" + result
    return result
```
